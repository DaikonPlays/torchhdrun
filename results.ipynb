```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("results/mnist.csv")
mnist_hdc_accuracy = data.accuracy.mean()
mnist_hdc_accuracy
```

```python
data = pd.read_csv("results/isolet.csv")
isolet_hdc_accuracy = data.accuracy.mean()
isolet_hdc_accuracy
```

```python
data = pd.read_csv("results/mnist-with-buffer.csv")
ax = sns.lineplot(x="epoch", y="accuracy", hue="resources", style="dataset", data=data)
ax.hlines(mnist_hdc_accuracy, 1, 10)
plt.plot()
```

```python
data = pd.read_csv("results/mnist-with-buffer.csv")
data = data[data.resources > 0]
ax = sns.lineplot(x="epoch", y="accuracy", hue="resources", style="dataset", data=data)
ax.hlines(mnist_hdc_accuracy, 1, 10)
plt.plot()
```

```python
data = pd.read_csv("results/isolet-with-buffer.csv")
ax = sns.lineplot(x="epoch", y="accuracy", hue="resources", style="dataset", data=data)
ax.hlines(isolet_hdc_accuracy, 1, 10)
plt.plot()
```
